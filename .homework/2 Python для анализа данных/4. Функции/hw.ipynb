```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def menu(user_input):
    '''Меню программы'''
    reply = [['Введите номер документа:',
             'Введите номер полки:',
             'Введите тип документа:',
             'Введите владельца документа:',
             'Введите полку для хранения:'],
             ['Владелец документа:',
             'Документ не найден в базе.',
             'Документ хранится на полке:',
             'Полка добавлена.',
             'Такая полка уже существует.',
             'Полка удалена.',
             'Текущий перечень полок:',
             'На полке есть документы, удалите их перед удалением полки.',
			       'Такой полки не существует.',
	      		 'Документ добавлен.',
	      		 'Добавьте полку командой as.',
	      		 'Документ удален.',
	      		 'Документ перемещен.',
             'Текущий список документов:']]
    if user_input == 'p':
        do = proprietor(input(f'{reply[0][0]} '))
        if len(do) > 0:
            print(f'{reply[1][0]} {do[0]}')
        else:
            print(f'{reply[1][1]}')
    if user_input == 's':
        sn = shelf(input(f'{reply[1][0]} '))
        if len(sn) > 0:
            print(f'{reply[1][2]} {sn[0]}')
        else:
            print(f'{reply[1][1]}')
    if user_input == 'l':
        listall()
    if user_input == 'ads':
        ns = input(f'{reply[0][1]} ')
        if ns in directories:
            print(f'{reply[1][4]} {reply[1][6]} {", ".join(directories.keys())}.') 
        else:
            add_shelf(ns)
            print(f'{reply[1][3]} {reply[1][6]} {", ".join(directories.keys())}.')
    if user_input == 'ds':
        sn = input(f'{reply[0][1]} ')
        if sn in directories.keys():
            if len(directories[sn]) > 0:
                print(f'{reply[1][7]} {reply[1][6]} {", ".join(directories.keys())}.') 
            else:
                del_shelf(sn)
                print(f'{reply[1][5]} {reply[1][6]} {", ".join(directories.keys())}.')
        else:
            print(f'{reply[1][8]} {reply[1][6]} {", ".join(directories.keys())}.') 
    if user_input == 'ad':
        dn, dt, do, ds = input(reply[0][0]), input(reply[0][2]), input(reply[0][3]), input(reply[0][4])
        if ds not in directories:
            print(f'{reply[1][8]} {reply[1][13]}')
            listall()
        else:
            add_document(dn, dt, do, ds)
            print(f'{reply[1][9]} {reply[1][13]}:')
            listall()
    if user_input == 'd':
        dn = input(f'{reply[0][0]}')
        if len(shelf(dn)) > 0:
            del_document(dn)
            print(f'{reply[1][11]}\n{reply[1][13]}')
            listall()
        else:
            print(f'{reply[1][1]}\n{reply[1][13]}')
            listall()
    if user_input == 'm':
        dn, ns = input(reply[0][0]), input(reply[0][1])
        if len(shelf(dn)) > 0 and ns in directories:
            mv_document(dn, ns)
            print(f'{reply[1][12]} {reply[1][13]}:')
            listall()
        elif len(shelf(dn)) > 0 and ns not in directories:
            print(f'{reply[1][8]} {reply[1][6]} {", ".join(directories.keys())}.')
        else:
            print(f'{reply[1][1]}\n{reply[1][13]}')
            listall()

def proprietor(dn):
    '''Функция для поиска владельца документа'''
    return [doc["name"] for doc in documents if dn == doc['number']]

def shelf(dn):
    '''Функция для поиска полки по номеру документа'''
    return [sn for sn, dl in directories.items() if dn in dl]

def listall():
    '''Функция для просмотра полной информации о документах'''
    for doc in documents:
        print(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {shelf(doc["number"])[0]}')

def add_shelf(ns):
    '''Функция для добавления полки'''
    directories[ns] = []

def del_shelf(sn):
    '''Функция для удаления полки'''
    del directories[sn]

def add_document(dn, dt, do, ds):
    '''Функция для добавления документов'''
    documents.append({'type': dt, 'number': dn, 'name': do})
    directories[ds].append(dn)

def del_document(dn):
    '''Функция для удаления документов'''
    documents.remove([doc for doc in documents if doc['number'] == dn][0])
    directories[shelf(dn)[0]].remove(dn)

def mv_document(dn, ns):
    '''Функция для перемещения документа между полками'''
    directories[shelf(dn)[0]].remove(dn)
    directories[ns].append(dn)

while True:
    user_input = input(f'Введите команду: ')
    if user_input == 'q':
        break
    menu(user_input)
```
