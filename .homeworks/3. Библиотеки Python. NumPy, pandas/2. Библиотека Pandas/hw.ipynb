### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера.  
Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
# импорт библиотеки pandas
import pandas as pd
```

```python
# импортируем данные из файлов (только необходимые столбцы)
movies = pd.read_csv('./movies.csv', usecols=['movieId', 'title'])
ratings = pd.read_csv('./ratings.csv', usecols=['movieId', 'rating'])
```

```python
# проверяем датафреймы
print(movies.head())
print(ratings.head())
```

```python
# фильтрация по оценке 5.0, группировка по movieId с подсчетом количества строк, возврат индекса с максимальным значением
id = ratings[ratings.rating == 5.0].groupby("movieId").count().idxmax()[0]
# фильтрация по полученному id
movies[movies.movieId == id]
```

### Задание 2  
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики  
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года.  
Не учитывайте в расчетах отрицательные значения quantity.

```python
# импортируем данные из csv
df = pd.read_csv('./power.csv')
# проверяем, что данные загружены
df.head()
```

```python
# список стран
countries = ['Latvia', 'Lithuania', 'Estonia']
# список категорий
categories = [4, 12, 21]
# период
start_date, end_date = 2005, 2010

# выстраиваем цепочку условий, выбираем столбец quantity и считаем по нему сумму
df[df.country.isin(countries) & df.category.isin(categories) & df.year.between(start_date, end_date) & df.quantity > 0]['quantity'].sum()
```
